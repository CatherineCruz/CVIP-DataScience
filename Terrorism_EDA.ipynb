# Exploratory Data Analysis of Global Terrorism

## <font color='blue'> Project Overview:

The Exploratory Data Analysis (EDA) of Global Terrorism Incidents project aims to analyze and gain insights from a comprehensive dataset on global terrorism incidents. The dataset provides information on various aspects of terrorism, including geographical locations, types of attacks, targets, and casualties. The primary goal is to explore patterns, trends, and factors associated with terrorism worldwide.




## <font color='blue'>Objectives:

**<font color='brown'>Data Exploration:</font>**
1. Gain a deep understanding of the dataset's structure and contents.
2. Identify key variables and their distributions.
3. Explore missing values, outliers, and data quality issues.

**<font color='brown'>Geospatial Analysis:</font>**
1. Visualize the geographical distribution of terrorism incidents on a world map.
2. Analyze hotspots and trends in different regions over time.
3. Investigate relationships between terrorism and geopolitical factors.

**<font color='brown'>Target Analysis:</font>**
1. Analyze the types of targets chosen by terrorists.
2. Explore changes in targeting preferences over time.
3. Identify high-profile targets or sectors frequently impacted.

**<font color='brown'>Casualty Analysis:</font>**
1. Investigate the number of casualties associated with different types of attacks.
2. Analyze trends in casualties over time and across regions.
3. Identify factors contributing to variations in casualty counts.

**<font color='brown'>Group and Perpetrator Analysis:</font>**
1. Examine the involvement of different terrorist groups.
2. Identify patterns in the tactics and strategies employed by various groups.
3. Explore the influence of specific groups on global terrorism trends.

```python

import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# Load world map with country boundaries
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))



# Merge with world map based on country name
merged_data = world.merge(top10_country, how='left', left_on='name', right_on='country_name')

# Plot the map
fig, ax = plt.subplots(1, 1, figsize=(15, 10))

# Plot all countries with a light gray outline
world.plot(ax=ax, color='lightgray', edgecolor='0.8')

# Highlight the top countries
merged_data.plot(column='count', cmap='plasma', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

ax.set_title('Countries with most attacks')
plt.show()
```

<div style="background-color: #E6DCEC; padding: 10px;">
    <h2><font color='black'>Time Line of number of people killed in the region "AMERICA" which includes <u>'North America', 'South America', 'Central America & Caribbean'</u></font></h2>
</div>



```python
# Creating a dataset region1
region1 = df[(df['region'] == 'North America') | (df['region'] == 'South America') | (df['region'] == 'Central America & Caribbean')]

region1
```

```python
# Creating a dataset region1_c

region1_c = region1.groupby(['country', 'year'])['killed'].size().reset_index()
region1_c
```

```python
plt.figure(figsize=(25,10))
sns.lineplot(x='year',y='killed',hue='country',data=region1_c,palette = 'magma_r')
plt.title("Total People killed in the 'AMERICA' region over years",fontsize=30)
plt.xlabel("Year",fontsize=20)
plt.ylabel("No of people killed",fontsize=20)
plt.grid(True)
plt.show()
```

```python
top_10_countries = region1_c.groupby('country')['killed'].sum().nlargest(10).index
top_10_df = region1_c[region1_c['country'].isin(top_10_countries)]

# Now you can use top_10_df for plotting
plt.figure(figsize=(10, 6))
for country, data in top_10_df.groupby('country'):
    plt.plot(data['year'], data['killed'], label=country)

plt.title('Top 10 Countries in the region America with maximum number of killings over years')
plt.xlabel('Year')
plt.ylabel('Number of Killed Incidents')
plt.legend()
plt.show()
```
